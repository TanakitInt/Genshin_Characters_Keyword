```python
import os
import sys
from collections import Counter
```

```python
file = open("Genshin_Impact_Characters_Keyword.txt", "r")
#print(f.read())

string_without_line_breaks = ""
for line in file:
  stripped_line = line.rstrip()
  string_without_line_breaks += stripped_line
file.close()

genshin_string = string_without_line_breaks
```

```python
# all word without newline
genshin_string
```

```python
# splitting into single word
genshin_list = genshin_string.split()

genshin_list
```

```python
# counting words
counters = Counter(genshin_list).most_common()

counters
```

```python
with open('genshin_counts.csv', 'w', newline='', encoding='utf-8') as csvfile:

    fieldnames = ['words', 'counts']
    writer = csv.writer(csvfile)
    writer.writerow(fieldnames)

    for key, value in counters:
        writer.writerow([key] + [value])
```
